# Perform Cat and Con Separation on X

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

# Create a Preprocession Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Num Pipeline
num_pipe =Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scaler',StandardScaler())])
```

```python
# Cat Pipeline
cat_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('ohe',OneHotEncoder(handle_unknown='ignore'))])
```

```python
# combine both num and cat pipe with ColumnTransformer
pre = ColumnTransformer([('num',num_pipe,con),('cat',cat_pipe,cat)]) 
```

```python
pre
```

# Preprocess the data

```python
X_pre = pre.fit_transform(X)
X_pre[0:5]
```

```python
cols=pre.get_feature_names_out()
cols
```

```python
X_pre= pd.DataFrame(X_pre,columns =cols)
X_pre.head()
```

# Apply train Test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest =train_test_split(X_pre,Y,test_size=0.2,random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

# Model Creation

```python
from sklearn.tree import DecisionTreeRegressor
model =DecisionTreeRegressor(max_depth=1,min_samples_leaf=6,min_samples_split= 5)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

# Hyperparameter tuning

```python
params ={'max_depth':[1,2,3,4,5,6,7,8],
         'min_samples_split':[5,6,7,8,9,10],
        'min_samples_leaf':[3,4,5,6,7,8],
         'criterion':['squared_error','absolute_error']}
```

```python
from sklearn.model_selection import GridSearchCV
dtr= DecisionTreeRegressor()
gscv = GridSearchCV(dtr,param_grid=params,cv=5,scoring='neg_mean_squared_error')
gscv.fit(xtrain,ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtr =gscv.best_estimator_
best_dtr
```

```python
best_dtr.score(xtrain,ytrain)
```

```python
best_dtr.score(xtest,ytest)
```

# Check Model mse, mae, rmse,r2

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
def evaluate_model(x,y,model):
    #predict train and test results
    ypred = model.predict(x)
    #evaluate the model
    mse =mean_squared_error(y,ypred)
    rmse =mse**(1/2)
    mae =mean_absolute_error(y,ypred)
    r2 =r2_score(y,ypred)
    # print the results
    print(f'Mean Squared Error :{mse:2f}')
    print(f'Root Mean Squared Error :{rmse :.2f}')
    print(f'Mean Absolute Error :{mae:2f}')
    print(f'R2 Score :{r2:.4f}')
```

```python
evaluate_model(xtrain,ytrain,best_dtr)
```

```python
evaluate_model(xtest,ytest,best_dtr)
```
